```python
# Basic import
import os
import sys
import json
import shutil
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
# Helper function
def writeProgress(msg, count, total):
    sys.stdout.write(msg + "{:.2%}\r".format(count/total))
    sys.stdout.flush()
    
def newPath(path):
    if not os.path.isdir(path):
        os.mkdir(path)

def read_json(src_path):
    with open(src_path, 'r') as json_file:
        data = json.load(json_file)
    return data

def write_json(data,dst_path):
    with open(dst_path, 'w') as outfile:
        json.dump(data, outfile)
```

```python
srcPath = './Following/'
fileList = os.listdir(srcPath)
print(len(fileList), fileList[:5])
```

```python
mlb = MultiLabelBinarizer()

accs = []
labels = []
for file in fileList:
    acc = file.split('.')[0]
    accs.append(acc)
    dic = read_json(srcPath + file)
    label = set([following['username'] for following in dic])
    labels.append(label)

arr = mlb.fit_transform(labels)
class_order = mlb.classes_
```

```python
df = pd.DataFrame(arr, columns = class_order)
df['movie'] = accs
df.set_index('movie', inplace = True)
df
```
