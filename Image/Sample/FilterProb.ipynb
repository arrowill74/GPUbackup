```python
# Basic import
import os
import sys
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_json(src_path):
    with open(src_path, 'r') as json_file:
        data = json.load(json_file)
    return data
```

```python
ig_pred = pd.read_csv('./output/resnet50_trainable_9e-5_newdata_IGphoto_result.csv')
ig_pred
```

```python
pred = ig_pred.iloc[:, 1:]
print(pred.shape, type(pred))
pred_np = pred.to_numpy()
print(pred_np.shape,type(pred_np))
```

```python
train_df = pd.read_csv('../filename2genreMat.csv')
train_df
```

```python
truth = train_df.iloc[:, 2:]
print(truth.shape, type(truth))
truth_np = truth.to_numpy()
print(truth_np.shape,type(truth_np))
print(truth_np[:1])
```

```python
sub = pred_np - truth_np
print(sub.shape, type(sub))
```

```python
threshold = 0.5
idxs = []
for i in range(len(sub)):
    if min(sub[i]) < -threshold:
        filename = ig_pred.loc[i, 'filename']
        print(i, min(sub[i]), filename)
#         img = plt.imread('../IGimg/' + filename)
#         plt.imshow(img)
#         plt.show()
    else:
        idxs.append(i)
```

```python
genres = read_json('../../orderedListGenres.json')
print(len(genres), genres)
```

```python
sub_df = pd.DataFrame(sub, columns=genres)
sub_df
```

```python
sub_df['filename'] = ig_pred['filename']
ordered_cols = ["filename"] + genres
sub_df = sub_df[ordered_cols]
sub_df
```

```python
len(idxs)
```

```python
filtered = sub_df.loc[idxs, :]
filtered
```

```python
filtered.to_csv('./output/filtered_5e-01.csv', index=False)
```
