```python
# Basic import
import os
import sys
import json
import shutil
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Helper function
def writeProgress(msg, count, total):
    sys.stdout.write(msg + "{:.2%}\r".format(count/total))
    sys.stdout.flush()
    
def newPath(path):
    if not os.path.isdir(path):
        os.mkdir(path)

def read_json(src_path):
    with open(src_path, 'r') as json_file:
        data = json.load(json_file)
    return data

def write_json(data,dst_path):
    with open(dst_path, 'w') as outfile:
        json.dump(data, outfile)
```

```python
PATH = '../Profiles/'
OUTPUT_PATH = './input/'
```

```python
genreDic = read_json('../genresDic.json')
len(genreDic)
```

```python
li = []
accs = os.listdir(PATH)
for acc in accs:
    accPath = PATH + acc
    
    desPath = accPath + '/description.json'
    data = read_json(desPath)
    text = data['summary']
    print(text)
    row = [acc, text, genreDic[acc]]
    li.append(row)
```

```python
print(len(li), li[:5])
```

```python
df = pd.DataFrame(li, columns = ['username', 'text', 'genres'])
df = df.sort_values(by = 'username')
df
```
