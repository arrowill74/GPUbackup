```python
import os
import json
import numpy as np
import pandas as pd
```

```python
# Helper function
def read_json(src_path):
    with open(src_path, 'r') as json_file:
        data = json.load(json_file)
    return data

def write_json(data,dst_path):
    with open(dst_path, 'w') as outfile:
        json.dump(data, outfile)
```

```python
columns = read_json('../orderedListGenres.json')
print(len(columns), columns)
```

```python
train_data_IG=pd.read_csv('./stopword/train_IG.csv')
train_data_IG = train_data_IG.dropna()
train_data_IG = train_data_IG.reset_index(drop=True)
train_data_IG
```

```python
movieCount = pd.DataFrame(train_data_IG.groupby(['username']).count())
movieCount = pd.DataFrame(movieCount['genres']).rename(columns={'genres':'count'})
movieCount
```

```python
movieCount.to_csv('./movieTextCount.csv')
```

```python
genredf = pd.read_csv('../genresMat.csv')
genredf = genredf.rename(columns={'movie':'username'})
genredf
```

```python
result = pd.merge(train_data_IG, genredf, on = 'username')
result = result.drop(columns = ['convert_text', 'genres'])
result
```
