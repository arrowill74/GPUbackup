```python
import os
import sys
import json
import numpy as np
import pandas as pd
```

```python
# Helper function
def writeProgress(msg, count, total):
    sys.stdout.write(msg + "{:.2%}\r".format(count/total))
    sys.stdout.flush()
    
def newPath(path):
    if not os.path.isdir(path):
        os.mkdir(path)

def read_json(src_path):
    with open(src_path, 'r') as json_file:
        data = json.load(json_file)
    return data

def write_json(data,dst_path):
    with open(dst_path, 'w') as outfile:
        json.dump(data, outfile)
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import *
import re

stpw = stopwords.words('english')
```

```python
username_title = pd.read_csv("./input/username_title_new.csv")
keys = username_title.username
values = username_title.title
compound = dict(zip(keys, values))
compound
```

```python
def textProcessing(string):
#     print(string)
    string = re.sub(r"http[\S]+", "", string)
    string = re.sub(r"www[\S]+", "", string)
#     print('drop hyperlink:', string)
    
    for key, v in compound.items():
        if not string.find(key) == -1:
#             print(key, 'exists')
            string = re.sub(key, ' '+v+' ', string)
    
    word_tokens = word_tokenize(string)
    word_tokens = [re.sub(r"[^a-z]+", "", t) for t in word_tokens] # filter non English words
    word_tokens = list(filter(None, word_tokens)) # filter empty string

    word_tokens = [w for w in word_tokens if not w in stpw] # remove stop word
    
#     stemmer = PorterStemmer()
#     words = [stemmer.stem(word) for word in word_tokens] # stemming
    
    words = list(filter(lambda x : len(x) > 1, word_tokens)) # filter empty string
    
    c = len(words)
    count.append(c)
    
    filtered_sentence = ' '.join(words)
    print(filtered_sentence)
    return filtered_sentence
```

# Training data

## IG

```python
train_data=pd.read_csv('./input/train_IG.csv')
train_data
```

```python
count = []
convert_text = []
for i, row in train_data.iterrows():
    string = row.text
    try:
        string = string.lower()
    except AttributeError:
        print(i, string)
        train_data = train_data.drop(i)
        continue

    convert_text.append(textProcessing(string))
```

```python
len(convert_text)
```

```python
train_data['convert_text'] = convert_text
train_data
```

```python
print('max:', max(count))
print('min:', min(count))
print('mean:', sum(count)/len(count))

counts = np.bincount(count)
print(np.argsort(counts)[::-1])
```

```python
train_data.to_csv('./stopword/train_IG.csv', index = 0, columns = ['username', 'convert_text', 'genres'])
train_data
```
