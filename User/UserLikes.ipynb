```python
# Basic import
import os
import sys
import json
import pandas as pd
import numpy as np
```

```python
def read_json(src_path):
    with open(src_path, 'r') as json_file:
        data = json.load(json_file)
    return data

def write_json(data,dst_path):
    with open(dst_path, 'w') as outfile:
        json.dump(data, outfile)
        
def writeProgress(msg, count, total):
    sys.stdout.write(msg + "{:.2%}\r".format(count/total))
    sys.stdout.flush()
```

```python
DIR = '../Posts_165/'

movie_order = read_json('../orderedListMovies.json')
print(len(movie_order), movie_order[:10])

user_order = read_json('../orderedListUsers.json')
print(len(user_order), user_order[:10])
```

```python
matrix = np.zeros((len(user_order), len(movie_order)))
matrix.shape
```

```python
for m in movie_order:
    print(movie_order.index(m), m)
    print('\tSum:', np.sum(matrix))
    c = 0
    for folder in os.listdir(DIR + m):
        try:
            file = read_json(DIR + m + '/' + folder + '/likes.json')
        except FileNotFoundError:
            continue
        writeProgress('Progress:', c, len(os.listdir(DIR + m)))
        c += 1
        for u in user_order:
            if u in file:
                matrix[user_order.index(u)][movie_order.index(m)] = matrix[user_order.index(u)][movie_order.index(m)] + 1
    print('\n\tSum:', np.sum(matrix))
```

```python
s = np.sum(matrix, axis = 1)
s.shape
```

```python
for i in s:
    if i < 20:
        print(i)
```
