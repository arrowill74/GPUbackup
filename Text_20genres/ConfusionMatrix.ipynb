```python
# Basic import
import os
import sys
import json
import shutil
import pandas as pd
import numpy as np
import numpy as np
from sklearn.metrics import multilabel_confusion_matrix
```

```python
# Helper function
def writeProgress(msg, count, total):
    sys.stdout.write(msg + "{:.2%}\r".format(count/total))
    sys.stdout.flush()
    
def newPath(path):
    if not os.path.isdir(path):
        os.mkdir(path)

def read_json(src_path):
    with open(src_path, 'r') as json_file:
        data = json.load(json_file)
    return data

def write_json(data,dst_path):
    with open(dst_path, 'w') as outfile:
        json.dump(data, outfile)
```

```python
genres = ['Biography',
         'Animation',
         'Sci-Fi',
         'Fantasy',
         'Romance',
         'Crime',
         'Mystery',
         'Horror',
         'Thriller',
         'Action',
         'Adventure',
         'Comedy',
         'Drama',
         'Others'] # 14 classes
others = ['Western', 'Documentary', 'War', 'History', 'Family', 'Sport', 'Music'] # 7 classes
```

```python
df = pd.read_csv('./output/result_word2vec_bi.csv')
df
```

```python
genresDic = read_json('genresDicOthers.json')
len(genresDic)
```

```python
li = []
for i, row in df.iterrows():
    acc = row['username']
    gs = genresDic[acc]
    print(acc, gs)
    glist = []
    newrow = [row['username']]
    for g in genres:
        if g in gs:
            glist.append(1)
        else:
            glist.append(0)
            
    li.append([acc] + glist)
```

```python
true_df = pd.DataFrame(li, columns = ['username'] + genres)
true_df.to_csv('./true_df.csv', index = 0)
true_df
```

```python
y_true = np.array(true_df.iloc[:, 1:])
y_pred = np.array(df.iloc[:, 4:])
print(y_true.shape, y_pred.shape)
```

```python
cmatrix = multilabel_confusion_matrix(y_true, y_pred)
cmatrix
```

```python
conmat = pd.DataFrame(data = {
    'genre' : genres,
    'TP' : cmatrix[:, 1, 1],
    'FP' : cmatrix[:, 0, 1],
    'FN' : cmatrix[:, 1, 0],
    'TN' : cmatrix[:, 0, 0]
}, columns=['genre', 'TP', 'FP', 'FN', 'TN'])
conmat
```
